### Venue Details
The code, below, finds the details on the Chicago venues.

```python
# Function to check if a json key exists and if so return its value.
def check_key(data, key):
    try:
        return data[key]
    except:
        return ""
```

```python
# Function to get food venue details from the Chicago venue id.
def get_venues_details(venue_id, client_id, client_secret, version):
    
    df = pd.DataFrame(columns=['id','phone','twitter','instagram','facebook','verified','checkins_count','users_count','tip_count','visits_count','url','price_tier','likes','rating','delivery_available','menu_available','specials_count','photos_count','here_now','listed','hours_days','attributes'])
    details_url = "https://api.foursquare.com/v2/venues/{}?&client_id={}&client_secret={}&v={}".format(venue_id, client_id, client_secret, version)
    results = requests.get(details_url).json()
    
    response = results['response']
    venue = response['venue']
    
    # get venue contact info
    contact = venue['contact']
    phone = check_key(contact,'phone')
    twitter = check_key(contact,'twitter')
    instagram = check_key(contact,'instagram')
    facebook = check_key(contact,'facebook')
    
    # verified
    verified = check_key(venue,'verified')
    
    # stats
    stats = venue['stats']
    checkins_count = check_key(stats,'checkingsCount')
    users_count = check_key(stats,'usersCount')
    tip_count = check_key(stats,'tipCount')
    visits_count = check_key(stats,'visitsCount')
    
    # url
    url = check_key(venue,'url')
    
    # price
    price_tier = check_key(venue['price'],'tier')
    
    # likes
    likes = check_key(venue['likes'],'count')
    
    # rating
    rating = check_key(venue,'rating')
    
    # delivery
    delivery = check_key(venue,'delivery')
    delivery_available = 0
    if delivery != "":
        if check_key(delivery,'id') != "":
            delivery_available = 1
        else:
            delivery_available = 0
        
    #menu
    menu = check_key(venue,'menu')
    menu_available = 0
    if menu != "":
        if check_key(menu,'url') != "":
            menu_available = 1
        else:
            menu_available = 0
      
    #specials
    specials = check_key(venue,'specials')
    specials_count = ''
    if specials != "":
        specials_count = check_key(specials,'count')
        
    # photos
    photos_count = check_key(venue['photos'], 'count')
    
    # here now
    here_now = check_key(venue['hereNow'], 'count')
    
    # listed
    listed = check_key(venue['listed'], 'count')
    
    # hours
    try:
        timeframes = check_key(venue['hours'],'timeframes')
        hours_days = ""
        for tf in timeframes:
            hours_days += check_key(tf,'days') + ": "
            hours_open = check_key(tf,'open')
            for ho in hours_open:
                hours_days += check_key(ho,'renderedTime') + ', '
        hours_days = hours_days[:-2]
    except:
        hours_days = "Couldn't get hours."
    
    # attributes
    attributes = venue['attributes']['groups']
    attrs = []
    for attr in attributes:
        items = attr['items']
        for item in items:
            attrs.append('{}: {}'.format(item['displayName'], item['displayValue']))
        
    # add data to dataframe
    df = df.append({'id':venue_id,'phone':phone,'twitter':twitter,'instagram':instagram,'facebook':facebook,'verified':verified,'checkins_count':checkins_count,'users_count':users_count,'tip_count':tip_count,'visits_count':visits_count,'url':url,'price_tier':price_tier,'likes':likes,'rating':rating,'delivery_available':delivery_available,'menu_available':menu_available,'specials_count':specials_count,'photos_count':photos_count,'here_now':here_now,'listed':listed,'hours_days':hours_days,'attributes':attrs}, ignore_index=True)
    return df
```

```python
# Read in chicago_food_venues.csv
chicago_food_venues = pd.read_csv('../saved_data/chicago_food_venues.csv')

# Initialize a dataframe to store Chicago food venue details.
chicago_food_venues_details = pd.DataFrame(columns=['id','phone','twitter','instagram','facebook','verified','checkins_count','users_count','tip_count','visits_count','url','price_tier','likes','rating','delivery_available','menu_available','specials_count','photos_count','here_now','listed','hours_days','attributes'])

# Loop over Chicago venues in chicago_food to get details of those venues.
for i in chicago_food_venues['id']:
    df = get_venues_details(i, CLIENT_ID, CLIENT_SECRET, VERSION)
    chicago_food_venues_details = chicago_food_venues_details.append(df, ignore_index=True)
chicago_food_venues_details
```

```python
# Split attributes column to get individual attributes and create dummies
# for those attributes.

# Find all unique attribute names.
attrs = []
for venue, attributes in zip(chicago_food_venues_details['id'], chicago_food_venues_details['attributes']):
    for attr in attributes:
        attr_name = attr.split(':')[0]
        if attr_name not in attrs:
            attrs.append(attr_name)

# Create dataframe columns for all unique attribute names.
for attr in attrs:
    chicago_food_venues_details[attr] = ''
    
# Fill in values of new dataframe columns from attributes column.
i = 0
for venue, attributes in zip(chicago_food_venues_details['id'], chicago_food_venues_details['attributes']):
    for attr in attributes:
        attr_name = attr.split(':')[0]
        attr_value = attr.split(':')[1].strip()
        chicago_food_venues_details.loc[i,attr_name] = attr_value
    i += 1

chicago_food_venues_details
            
```

```python
# Save intermediate dataset.
chicago_food_venues_details.to_csv("../saved_data/chicago_food_venues_details_messy.csv")
```

```python
# Create dummy variables to indicate whether a Chicago food venue has certain attributes.
# Blank entries mean that the food venue did not have details on the given attribute.
# This does not necessarily mean that the food venue does not have this attribute but just that they
# did not post about.
# This is as good as not having it, though, because its advertised attributes that draw in customers.
details_cleaned = chicago_food_venues_details

cols = ['phone','twitter','instagram','facebook','url']
for col in cols:
    details_cleaned[col + '_available'] = np.where(details_cleaned[col] != "", 1, 0)

details_cleaned['verified'] = np.where(details_cleaned['verified'] == True, 1, 0)

cols = ['Credit Cards','Outdoor Seating','Wi-Fi','Reservations','ATM','Restroom','Wheelchair Accessible','TVs','Private Room','Music','Coat Check']
for col in cols:
    try:
        details_cleaned[col.lower()] = np.where((details_cleaned[col].str.contains('Yes') | details_cleaned[col].str.contains('Free')), 1, 0)
    except:
        print(col + " not in dataframe")
cols = ['Breakfast','Dessert','Happy Hour','Lunch','Brunch','Dinner','Beer','Wine','Full Bar','Cocktails','Street Parking','Public Lot','Live Music','Jukebox','Bar Service', 'Bar Snacks', 'Tasting Menu', 'Delivery','Street Parking','Valet Parking','BYO','Parking','Table Service','Public Lot']
for col in cols:
    try:
        details_cleaned[col.lower()] = np.where(details_cleaned[col] != "", 1, 0)
    except:
        print(col + " not in dataframe")
        
details_cleaned
```

```python
# Drop variables for which we made dummies
details_cleaned['price'] = np.where(details_cleaned['Price']=='$$$',3,4)
details_cleaned = details_cleaned.drop(columns=['phone','twitter','instagram','facebook','url','attributes','checkins_count','users_count','visits_count','Price','Credit Cards','Outdoor Seating','Wi-Fi','Reservations','ATM','Restroom','Wheelchair Accessible','TVs','Private Room','Music','Coat Check','Breakfast','Dessert','Happy Hour','Lunch','Brunch','Dinner','Beer','Wine','Full Bar','Cocktails','Street Parking','Public Lot','Live Music','Jukebox','Bar Service', 'Bar Snacks', 'Tasting Menu', 'Delivery','Street Parking','Valet Parking','BYO','Parking','Table Service','Public Lot'])
```

```python
# Merge on details to food venues
chicago_food_venues_w_details = chicago_food_venues.merge(details_cleaned, how='left', on='id')
chicago_food_venues_w_details = chicago_food_venues_w_details.drop(columns=['Unnamed: 0'])
chicago_food_venues_w_details
```

```python
# Save the data to a csv so we do not need to run the code each time we start the program.
# This will save the number of API calls we need to make.
chicago_food_venues_w_details.to_csv('../saved_data/chicago_food_venues_details_cleaned.csv')
```
